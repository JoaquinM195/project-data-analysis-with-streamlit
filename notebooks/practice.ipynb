```python
import pandas as pd
import numpy as np

# Cargar el dataset Boston desde un CSV equivalente (Boston original)
url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"

df_boston = pd.read_csv(url)
df_boston.columns = df_boston.columns.str.upper()
```

```python
crim_min = df_boston["CRIM"].min()
crim_max = df_boston["CRIM"].max()

crim_min, crim_max
```

```python
# valor máximo de CRIM "elegido por el usuario" (después será el slider)
max_crim = 5   # probá cambiarlo a otros valores

# valor del checkbox "Limita con Río Charles (CHAS)"
only_river = True   # poné False para probar sin filtro por río
```

```python
# Filtro por CRIM
df_filtrado = df_boston[df_boston["CRIM"] <= max_crim].copy()

# Si el checkbox está tildado, filtramos también por CHAS = 1
if only_river:
    df_filtrado = df_filtrado[df_filtrado["CHAS"] == 1]

df_filtrado.head(), df_filtrado.shape
```

```python
medv_values = df_filtrado["MEDV"].values  # array de NumPy

mean_medv = np.mean(medv_values)
median_medv = np.median(medv_values)
std_medv = np.std(medv_values, ddof=1)  # desviación estándar muestral

mean_medv, median_medv, std_medv
```

```python
print(f"Media de MEDV: {mean_medv:.2f}")
print(f"Mediana de MEDV: {median_medv:.2f}")
print(f"Desviación estándar de MEDV: {std_medv:.2f}")
```
